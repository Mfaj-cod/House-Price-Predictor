#### House price prediction


```python
import pandas as pd
import pickle
```

```python
df = pd.read_csv('data\\house_price_regression_dataset.csv', index_col=False)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
df.drop('Num_Bathrooms', axis=1, inplace=True)
```

```python
df.drop('Neighborhood_Quality', axis=1, inplace=True)
```

```python
x = df.drop('House_Price', axis=1)
y = df['House_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
with open('resources\\scaler.pkl', 'wb') as s:
    pickle.dump(scaler, s)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_true=y_test, y_pred=y_predicted)
r2 = r2_score(y_pred=y_predicted, y_true=y_test)

print(f"mean squared error: {mse}")
print(f"r2Score: {r2}")
```

```python
import pickle

with open('resources\\model.pkl', 'wb') as m:
    pickle.dump(model, file=m)
```
